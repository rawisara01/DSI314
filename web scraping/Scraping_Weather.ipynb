```python
from datetime import datetime
import openpyxl
import requests
from tabulate import tabulate
import pandas as pd

page = 20190101
Time = []
Temperature =[]
Dew_Point =[]
Humidity =[]
Wind =[]
Wind_Speed=[]
Wind_Gust=[]
Pressure=[]
Precip=[]
Condition=[]
while page <=20190131:
    response = requests.get('https://api.weather.com/v1/location/VTBD:9:TH/observations/historical.json?apiKey=e1f10a1e78da46f5b10a1e78da96f525&units=m&startDate='+str(page)).json()["observations"]
    weather_data = sorted(response, key=lambda k: k["valid_time_gmt"])

    header = [
        "Time", "Temperature", "Dew Point", "Humidity", "Wind",
        "Wind Speed", "Wind Gust", "Pressure", "Percip.", "Conditions",
    ]

    for item in weather_data:
        Time.append(datetime.fromtimestamp(item["valid_time_gmt"]).strftime('%I:%M %p'))
        Temperature.append(f'{item["temp"]} °C')
        Dew_Point.append(f'{item["dewPt"]} °C')
        Humidity.append(f'{item["rh"]} %')
        Wind.append(item["wdir_cardinal"])
        Wind_Speed.append(f'{item["wspd"]} km/h')
        Wind_Gust.append(f'{item["gust"] if item["gust"] else 0} km/h')
        Pressure.append(f'{item["pressure"]} hPa')
        Precip.append(f'{item["precip_total"] if item["precip_total"] else "0.0 mm"}')
        Condition.append(item["wx_phrase"])
    print('Complete page:',page)
    page += 1
weather = pd.DataFrame([Time,Temperature,Dew_Point,Humidity,Wind,Wind_Speed,Wind_Gust,Pressure,Precip,Condition]).transpose()
```

```python
weather.columns = ['Time','Temperature','Dew_Point','Humidity','Wind','Wind_Speed','Wind_Gust','Pressure','Precip','Condition']
```

```python
weather.set_index('Time')
```

```python
weather.to_excel('2019.xlsx',engine='openpyxl')
```
